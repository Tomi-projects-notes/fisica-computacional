```python
### integral de Simposn
def f(x):
    return 0

def integral_Simpson (a,b,N):
    h = (b-a)/N
    s = f(a)*h/3+f(b)*h/3
    for k in range(1,N): 
        if k%2==0:
            s+=f(a+k*h)*h*2/3
        else:
            s+=f(a+k*h)*h*4/3
    return s

#mais rápido
def integral_Simpson2(a,b,N):
    h = (b-a)/N
    s=0
    for k in range(1,N,2):
        s+=f(a+k*h)*4+f(a+(k+1)*h)+f(a+(k-1)*h)
    return s

### Simpson adaptativo
def integral_Simpson_adaptativo(a,b,N,e):
    h=(b-a)/N
    t=0
    s=(f(a)+f(b))*(1/3)
    for i in range (2,N-1,2):
        s+=f(a+i*h)*(2/3)
    for i in range (1,N,2):
        t+=f(a+i*h)*(2/3)
    r=h*(s+2*t)
    c=e+1
    n=1
    while abs(c)>abs(e):
        N*=2
        c=-r
        h=h/2
        s+=t
        t=0
        for i in range (1,N,2):
            t+=f(a+i*h)*(2/3)
        r=(s+2*t)*h
        c+=r
        c*=(1/15)
        n+=1
    return r
```

```python
### integral método do trapézio
def g(x):
    return 0

def integral_trapézio1(a,b,N):
    h=(b-a)/N
    r=0
    for i in range (1,N):
        r+= (g(a+(i-1)*h)+g(a+(i*h)))*h/2
    return r

#mais rápido
def integral_trapézio2 (a,b,N):
    h=(b-a)/N
    r=(g(a)+g(b))*h/2
    for i in range (1,N-1):
        r+=g(a+i*h)*h
    return r

def integral_trapézio_adaptativo(a,b,N,e):
    h=(b-a)/N
    r=(g(a)+g(b))*0.5
    s=1
    for i in range (1,N-1):
        r+=g(a+i*h)
    n=1
    while abs(s)>abs(e) :
        s=-r*h
        N*=2
        h=(b-a)/N
        for i in range(1,N-1,2):
            r+=g(a+i*h)
        s+=r*h
        s*=(1/4**n-1)
        n+=1
    return r*h
```

```python
### integral de Romberg
def p(x):
    return 0

def integral_Romberg(a,b,N,e):
    matriz=np.zeros((6,6))
    h=(b-a)/N
    r=(p(a)+p(b))*(h/2)
    for i in range (1,N-1):
        r+=p(a+i*h)*h
    matriz[0][0]=r
    s=e+1
    l=1
    while abs(s)>abs(e):
        h=h/2
        N*=2
        r=(p(a)+p(b))*(h/2)
        for i in range (1,N-1):
            r+=p(a+i*h)*h
        matriz[l][0]=r
        for j in range (1,l+1):
            s=(matriz[l][j-1]-matriz[l-1][j-1])/(2**j-1)
            matriz[l][j]= matriz[l][j-1]+s
        l+=1
    return (matriz)
```

```python
### quadraturas
def f(x):
    return 0

# as mudanças de variável podem não estar certas

def quadratura_gaussiana(a,b,N):
    x, w = np.polynomial.legendre.leggauss(N)
    xx=0.5 * (b - a) * x + 0.5 * (b + a)
    ww=0.5 * (b - a) * w
    return np.sum(ww * f(xx))

def quad_hermite(a,b,n):
    zeros, pesos = roots_hermitenorm(n)
    return  np.sum(f(zeros) * pesos)

def quad_chebyshev(a,b,n):
    x, w = roots_chebyt(n)
    xx=0.5 * (b - a) * x + 0.5 * (b + a)
    ww=0.5 * (b - a) * w
    def g(x):
        return f(xx)*(np.sqrt(1-x**2))
    return np.sum(g(xx) * ww)

def legendre_quadrature(a,b, n):
    x, w = roots_legendre(n)
    xx=0.5 * (b - a) * x + 0.5 * (b + a)
    ww=0.5 * (b - a) * w 
    return np.sum(ww * f(xx))

def quadratura_Laguerre(a:0,b:+inf,N):
    x, w = roots_laguerre(N)
    return f(x)*w
```

```python
#inversa de uma matriz
A=np.array(,float)

def inversa(A):
    v=np.identity(len(A))
    for m in range(len(A)): 
        div=A[m,m]
        A [m, : ] /= div 
        v[m,:] /= div
        for i in range(m+1,len(v)): 
            mult=A [i ,m]
            A[i,:] -= mult*A[m,:] 
            v[i,:] -= mult*v[m,:] 
    return v

###decomposição LU por eliminação de Gauss
def U_L_Gauss(A):
    I=np.identity(len(A))
    for m in range(len(A)): 
        div=A[m,m]
        A [m, : ] /= div 
        I[m,:] /= div
        for i in range(m+1,len(v)): 
            mult=A [i ,m]
            A[i,:] -= mult*A[m,:] 
            I[i,:] -= mult*I[m,:]
        if m==0:
            r=inversa(I)
        else:
            r=np.dot(r,inversa(I))
    ### A=U e r=L
    return A,r
```

```python
### decomposição LU pelo método de Carot
A=np.array(,float)
v=np.array(,float)

def L_U_Carot(A):
    U=np.identity(len(A))
    L=np.zeros((len(A),len(A)))
    for i in range (len(A)):
        for j in range (i,len(A)):
            L[j,i]=A[j,i]
            for k in range (i):
                L[j,i]-=L[j,k]*U[k,i]
            U[i,j]=A[i,j]
            for l in range (i):
                U[i,j]-=L[i,l]*U[l,j]
            U[i,j]/=L[i,i]
    return L,U

def L_U_Carot_pivotagem_parcial(A):
    U=np.identity(len(A))
    L=np.zeros((len(A),len(A)))
    linha=np.arange(len(A))
    coluna=np.arange(len(A))
    for i in range (len(A)):
        a=i+np.argmax(A [i:,i])
        b=linha[a]
        if linha[i]!=linha[a]:
            linha[a]=linha[i]
            linha[i]=b
        for j in range (i,len(A)):
            L[linha[j],coluna[i]]=A[linha[j],i]
            for k in range (i):
                L[linha[j],coluna[i]]-=L[linha[j],coluna[k]]*U[k,i]
            U[i,j]=A[linha[i],j]
            for l in range (i):
                U[i,j]-=L[linha[i],l]*U[l,j]
            U[i,j]/=L[linha[i],coluna[i]]
    return L,U,linha
```

```python
#para resolver o sistema de equações com decomposição LU
#correr célula anterior

A=np.array(,float)
v=np.array(,float)

def matriz_y(v,L):
    y=np.zeros(len(v))
    for m in range(len(v)):
        y[m]=v[m]
        for k in range(0,m):
            y[m]-=L[m,k]*y[k]
        y[m]/=L[m,m]
    return y

def se_decomposiçãoLU(A,v):
    L,U,linha=L_U_Carot_pivotagem_parcial(A)
    y=matriz_y(v[linha],L[linha])
    xx = np.zeros(len(v))
    xx[-1] = y[-1] / U[len(v)-1, len(v)-1]
    for i in range(len(v)-1, -1, -1):
        xx[i] = (y[i] - np.dot(U[i][i+1:],xx[i+1:])) / U[i][i]
    return xx
```

```python
#decomposição QR

A=np.array(,float)

def decomp_QR_Gram_shimidt(A):
    Q=np.zeros((len(A),len(A)),float)
    R=np.zeros((len(A),len(A)),float)
    for i in range (len(A)):
        Q[:,i]=A[:,i]
        for j in range (0,i):
             Q[:,i]-=np.dot(Q[:,j],A[:,i])*Q[:,j]
        Q[:,i]/=np.linalg.norm( Q[:,i])
    R=np.dot(Q.transpose(),A)
    return Q,R

def decomp_QR_Householder(A):
    Q=np.identity(len(A),float)
    for j in range(len(A)):
        x=A[j:,j].copy() #ai
        e=np.zeros(len(x))
        e[0]=np.sign(x[0]) #ei
        v=np.linalg.norm(x)*e+x #ai+-||x||ei
        v/=np.linalg.norm(v)
        Q[j:,:]-=2*np.outer(v,np.dot(v,Q[j:,:])) # I-2v(v.Q)
    R=np.dot(Q.transpose(),A)
    return Q,R
```

```python
def f(x):
    return 0

def metodo_bissecção(a,b,erro):
    c=(a+b)/2
    e=erro+1
    fa=f(a)
    fb=f(b)
    while erro<abs(e):
        fc=f(c)
        if fc<=1e-16:
            return c
        elif fa*fc<0:
            b=c
            fb=fc
        elif fb*fc<0:
            a=c
            fa=fc
        else:
            raise ValueError("O sinal de f é o mesmo em a e em b.")
        c=(a+b)/2
        e=(b-a)/2
    return c
```

```python
def f(x):
    return 0

#está feita para encontrar um mínimo se quiser determinar um máximo fazer -f(x)
def metodo_sec_aurea(x1,x4,erro):
    z=(1+np.sqrt(5))/2
    x2=x4-(1/z)*(x4-x1)
    x3=x1+(1/z)*(x4-x1)
    e=erro+1
    while erro<e:
        if f(x2)<f(x3):
            x4=x3
            x3=x2
            x2=x4-(1/z)*(x4-x1)
        else:
            x1=x2
            x2=x3
            x3=x1+(1/z)*(x4-x1)
        e=x4-x1
    return (x2+x3)/2
```

```python
#método de Gauss_Newton

def P(x):
    return 0
h=0.01

def derivada_P(x,h):
    return (P(x+h/2)-P(x-h/2))/h

def segunda_derivada_P(x,h):
    return  (P(x+h)-2*P(x)+P(x-h))/h**2

def método_Gauss_Newton(x0,erro):
    e=erro+1
    while erro<e:
        x=x0-derivada_P(x0,h)/segunda_derivada_P(x0,h)
        e=x-x0
        x0=x
    return x0
```

```python
### Gauss_Seidel

A = np.array([[,],[,]],float) 
v = np.array([,],float)

def gauss_seided(A,v,erro,x0):
    N = len(A)
    D= np.linalg.inv(A*np.identity(len(A)))
    B = np.identity(len(A)) - np.dot(D,A)
    d=np.dot(D,v)
    e = 1+erro
    x=np.zeros(N)
    while (e >= erro):
        for i in range(N):
            x[i] = np.dot(B[i,0:],x) + d[i]
        e = np.max(np.abs(x - x0))/np.max(np.abs(x))
        x0 =np.copy(x)
    return x
```
