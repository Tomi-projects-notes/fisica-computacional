```python
#Ex 4 - calculo de um integral de um semicirculo


def integral(N):
    I=0
    h=2/N
    for k in range(N):
        y_k=math.sqrt(1-(-1+h*k)**2)
        I+=h*y_k
    return I
print(integral(1000))

%timeit -r 7 -n 1000 integral(10000)
```

```python
#Ex 2 - Integral de [0,2] da função x^4 -2x + 1
def g(x):
    return x**4 - 2*x + 1

def simpson(g, a , b , N):
    x = np.linspace(a,b,N+1)
    y = g(x)
    I=(1/3)*((b-a)/(N)*( y[0] + y[N] + 4*np.sum(y[1:N:2]) + 2*np.sum(y[2:N-1:2])))
    return I

def trape(g, a ,b , N):
    x = np.linspace(a,b,N+1)
    y = g(x)
    I=((b-a)/N)*( y[0] + y[N] + np.sum(y[0:N+1]))
    return I

I1 = simpson(g,0,2,1000)
I2 = trape(g,0,2,1000)
I_exato = 4.4

erro1 = abs(I_exato - I1)/I_exato
erro2 = abs(I_exato - I2)/I_exato

print(I1)
print("Erro da regra Simpson: ", erro1, "\n")

print(I2)
print("Erro da regra do Trapézio: ", erro2,"\n" )

%timeit -r 7 -n 1000 simpson(g,0,2,1000)
%timeit -r 7 -n 1000 trape(g,0,2,1000)
```
