### Erro numérico
Os números float apenas podem ser representados com 16 algarismos significativos de precisão

Além disso, ter cuidado com a igualdade entre floats 3.3 é diferente de 3.299999


---
Lembrar que x tem 16 algarismos significativos (precisos), logo devemos ter um erro numérico (epsilon) tal que:
\begin{align}
\epsilon = x \times 10^{-16}
\end{align}

Assim, sendo $C = 10^{-16}$ é a *constante de erro*

E a *variância* será então: $\sigma = Cx$

---

Dadas duas variáveis somadas ou subtraídas, $x_1$ e $x_2$, com variâncias $\sigma_1$ $\sigma_2$, então temos que:

\begin{align}
\\
\sigma^{2} = \sigma_1^{2} + \sigma_2^{2}
\\
\\
\Leftrightarrow \sigma = \sqrt{\sigma_1^{2} + \sigma_2^{2}}
\\
\\
\Leftrightarrow \sigma = \sqrt{C_1^{2} x_1^{2} + C_2^{2} x_2^{2}}
\end{align}

Já para o produto ou divisão de  duas variáveis, $x_1$ e $x_2$, tal que $x=x_1 x_2$ ou $x = \frac{x_1}{x_2}$ é:

\
\begin{align}
\sigma = \sqrt{2}Cx
\end{align}

Para uma soma $N$ variáveis, ${x_1,...,x_N}$ com variância, $\sigma_i = Cx_i$, então a variância total será:

\begin{align}
\sigma^2 = \sum_{i=1}^{N} \sigma_i^2 = \sum{i=1}^{N} C^2 x_i^2 = N C^2 \overline{x^2}
\end{align}

\begin{align}
\sigma = \sqrt{N} C \sqrt{\overline{x^2}}
\end{align}

Nota: $\sigma$ aumenta com o número de elementos combinados, $N$ elementos.

---

```python
#Se eu obter num cálculo 3.29999999999
x=3.2999999999

#Para verificar que x = 3.3

epsilon = 1e-6 # erro de arredondamento

if abs(x-3.3)<epsilon:
    print("x é igual a 3.3")
```

### Regra do Trapézio
A área da região abaixo dos pontos $f(x)$, tal que em $x \in [a,b]$ da função $f$ é dada por:

\
\begin{align}
I(a,b) = \int_{a}^{b} f(x) \, dx
\end{align}

---
Este é o valor exato da área nessa região. No entanto, computacionalmente este integral apenas pode ser resolvido por métodos numéricos como a *regra do trapézio*.

Primeiramente dividimo o intervalo $[a,b]$ em $N$ intervalos com um espaçamento $h= \frac{(b-a)}{N}$.

Assim sendo, teremos $N$ trapézios, sendo a área cada um deles dada por:

\
\begin{align}
A_k = \frac{h}{2}\left[f(a+(k-1)h) + f(a+kh)\right]
\end{align}

Nota: numeramos os trapézios por índices $k \in \{1,2,...,N\}$

Logo, podemos aproximar o integral de $a$ a $b$ por:
\
\begin{align}
\\
I(a,b) &\approx \sum_{k=1}^{N}A_k \\
&= \frac{h}{2} \sum_{k=1}^{N}\left[f(a+(k-1)h) + f(a+kh)\right] \\
&= h\left[\frac{1}{2}f(a) + f(a+h) + f(a+2h) + ... + \frac{1}{2}f(b)\right] \\
&= h\left[\frac{1}{2}f(a) + \frac{1}{2}f(b) + \sum_{k=1}^{N-1}f(a+kh)\right]
\end{align}

---


```python
# Algoritmo ideal para Regra do Trapézioo

def Trapezio(f,a,b,N): #f é a função que desejamos integrar
    h = (b-a)/N
    x = np.linspace(a,b,N)
    y = f(x)
    I = (h/2)*(y[0] + y[N-1] + 2*np.sum(y[1:N]))
    return I

#Exemplo:
def f(x):
    return x**4 - 2*x + 1
#Resultado:
print(Trapezio(f,0,2,10000))
```

### Regra de Simpson
Permite-nos calcular a área da região de maneira mais precisa, através de curvas quadráticas. No entanto, importante realçar que uma curva deste tipo necessita de 3 pontos, por isso vamos considerar que estes pontos estão espaçado por $h$.

Vamos ter os pontos $x_1=-h$, $x_2=0$ e $x_3=h$, a curva quadrática deverá seguir a forma: $Ax^2 + Bx + C$; e os valores de $f(x)$ serão:

\
\begin{align}
&f(-h) = Ah^2 - B h +C\\ \\&f(0) = C\\ \\&f(h) = Ah^2 + Bh + C      
\end{align}

Podemos concluir, então:

\
\begin{align}
&A = \frac{1}{h^2} \left[\frac{1}{2}f(-h) - f(0) + \frac{1}{2}f(h)\right] \\ \\&C = f(0)\\ \\&B = \frac{1}{2h} \left[f(h)-f(-h)\right]      
\end{align}

Nisto, a área de cada curva quadrática será dada por:

\
\begin{align}
\int_{-h}^{h} (Ax^2 + Bx + C)\, dx = \frac{2}{3}Ah^3 + 2Ch = \frac{1}{3}h \left[f(-h) + 4f(0) + f(h) \right]
\end{align}

Agora para obtermos o $I(a,b)$, necessitamos de somar todas as áreas das curvas quadráticas defenidas por 3 pontos:

\
\begin{align}
I(a,b) = \frac{h}{3} ( \left[ f(a) + 4f(a+h) + f(a+2h) \right] + \left[ f(a+2h) + 4f(a+3h) + f(a+4h) \right] \\ +   (...)    + \left[ f(a + (N-2)h) + 4f(a+(N-1)h) + f(b) \right] )
\end{align}

\
\begin{align}
I(a,b) &\approx \frac{h}{3} \left[ f(a) + 4f(a+h) + 2f(a+2h) + 4f(a+3h) + ... + f(b) \right]\\ &= \frac{h}{3} \left[ f(a) + f(b) + 4 \sum_{k\:ímpar}^{N-1} f(a+kh) +  2 \sum_{k\:par}^{N-2} f(a+kh) \right]
\end{align}

---

```python
#Algoritmo ideal para a Regra de Simpson

def Simpson(f,a,b,N):
    h=(b-a)/N
    x=np.linspace(a,b,N+1)
    y=f(x)
    I= h/3 * (y[0] + y[N] + 4*np.sum(y[1:N:2]) + 2*np.sum(y[2:N:2]))
    return I

#Exemplo:
def f(x):
    return x**4 - 2*x + 1
#Resultado:
print(Simpson(f,0,2,1000))
```

### Erro de aproximação 
Consideremos o integral de $x_k-1$ a $x_k$, expandindo $f(x)$ em série de Taylor, em torno do ponto $x_{k-1}$:

\
\begin{align}
f(x) = f(x_{k-1})+f'(x_{k-1})(x - x_{k-1}) + \frac{f''(x_{k-1}}{2!}(x - x_{k-1})^2
\end{align}

Assim, podemos calcular o integral a partir da série de Taylor:

\
\begin{align}
\int_{x_{k-1}}^{x_k} f(x) \, d(x) = f(x_{k-1})\int_{x_{k-1}}^{x_k}\,dx + f'(x_{k-1})\int_{x_{k-1}}^{x_k}(x - x_{k-1})\,dx + ...
\end{align}

Substituindo $u =x_k - x_{k-1}$:

\
\begin{align}
\int_{x_{k-1}}^{x_k} f(x) \, d(x) &= f(x_{k-1})\int_{0}^{h}\,du + f'(x_{k-1})\int_{0}^{h}u\,du + ...\\ &= hf(x_{k-1}) + \frac{h^2}{2}f'(x_{k-1}) + \frac{h^3}{6}f''(x_{k-1}) + O(h^4)
\end{align}


Podemos aplicar a integração a uma série de Taylor, desta vez em torno de $x_k$:

\
\begin{align}
\int_{x_{k-1}}^{x_k} f(x) \, d(x) &= hf(x_k) - \frac{h^2}{2}f'(x_k) + \frac{h^3}{6}f''(x_k) - O(h^4)
\end{align}

Considerando a média da integração das duas expansões, temos:

\
\begin{align}
\int_{x_{k-1}}^{x_k} f(x) \, d(x) &= \frac{h}{2} \left[ f(x_{k-1}) + f(x_k) \right] + \frac{h^2}{4} \left[ f'(x_{k-1}) - f'(x_k) \right] + \frac{h^3}{12} \left[ f''(x_{k-1}) + f''(x_k) \right] + O(h^4)
\end{align}

Agora somamos todas as fatias $k$ para obter o valor total do integral:

\
\begin{align}
\int_{a}^{b}f(x)\, dx &= \sum_{k=1}^{N} \int_{x_{k-1}}^{x_k} f(x)\, dx
\\ 
&= \frac{h}{2}\sum_{k=1}{N} \left[ f(x_{k-1}) + f(x_k) \right] + \frac{h^2}{4} \left[ f'(a) - f'(b) \right] + O(h^3)
\\
&= I_{T}(a,b) + O(h^3)
\end{align}

#### Nota 1:
A primeira soma do lado direito é equivalente à regra do trapézio.
#### Nota 2:
O resto da soma $O(h^3)$ é a quantidade necessária adicionar ao $I_t$ para obtermos o valor exato de $I(a,b)$, ou seja, é o erro de aproximação em relação ao uso da regra do trapézio.


Se substituirmos $f(x) \rightarrow f''(x)$, temos:

\
\begin{align}
\int_{a}^{b}f''(x)\, dx = \frac{h^2}{2} \sum_{k=1}^{N} \left[ f''(x_{k-1}) + f''(x_k) \right] + O(h^2)
\end{align}
\
Multiplicando por $\frac{h^2}{6}$, temos:

\
\begin{align}
\frac{h^2}{6} \int_{a}^{b} f''(x) \, dx + O(h^4) &= \frac{h^3}{12} \sum_{k=1}^{N} \left[ f''(x_{k-1}) + f''(x_k) \right]\\ &= \frac{h^2}{6} \left[ f'(b) - f'(a) \right] + O(h^4)
\end{align}

Conclusão:

\begin{align}
\int_{a}^{b} f(x) \, dx = \frac{h}{2} \sum_{k=1}^{N} \left[ f(x_{k-1}) + f(x_k) \right] + \frac{h^2}{12} \left[ f'(a) - f'(b) \right] + O(h^4)
\end{align}

E como: 

\begin{align}
I_T = \frac{h}{2} \sum_{k=1}^{N} \left[ f(x_{k-1}) + f(x_k) \right]
\end{align}

Então o erro aproximado que cometemos usando a regra do trapézio é:

\
\begin{align}
\epsilon = \frac{h^2}{12} \left[ f'(a) - f'(b) \right] + O(h^4)
\end{align}

Nota: não consideramos termos de maior ordem, ou seja, normalmente descarta-se $O(h^4)$

Além disto, temos o erro de arredondamento que terá um valor na ordem da constante $C=10^{-16}$ em Python

### Erro para a regra do trapézio
Na teoria poderiamos concluir que quanto menor o valor de $h$ (espaçamento entre trapézios), ou por outras palavras, quanto maior o valor de $N$ (números de intervalos), menor será o erro que iremos cometer e maior será a precisão da nossa integração numérica, mas isso não é verdade, pois quando:

\
\begin{align}
\epsilon = \frac{h^2}{12} \left[ f'(a) - f'(b) \right] \approx C \int_{a}{b}f(x) \, dx
\end{align}

Ou seja, se $diminuirmos\ h \implies \epsilon < C \int_{a}^{b}f(x) \, dx$, e portanto não iremos obter mais precisão e o nosso código será mais lento.

\
\begin{align}
h &= \sqrt{\frac{12 \int_{a}^{b} f(x) \, dx}{f'(a) - f'(b)}C}
\\
\\
N &= (b-a) \sqrt{{f'(a) - f'(b)}\frac{12 \int_{a}^{b} f(x) \, dx}C}
\end{align}




### Erro para a regra de Simpson

\\

### Forma prática de estimar erros
Consideremos que vamos integrar em $x \in [a,b]$ e vamos usar a Regra do Trapézio, recorrendo a $N_1$ intervalos com $h_1 = (b-a) /N_1$.

\
Agora consideremos que faremos a mesma integração mas agora em $N_2$ intervalos, tal que $N_2 = 2N_1$, com $h_2=(b-a)/N_2 \Rightarrow h_2 = 2h_1$.

\
É importante relembrar que a Regra do Trapézio introduz um erro de $O(h^2)$, o que significa que se tivermos $\frac{h}{2}$, então temos um erro de $\frac{O (h^2)}{4}$. Ou seja, facilmente verficamos que: maior o número de intervalos $N$ $\Rightarrow$ menor $h$ $\Rightarrow$ maior erro.

\

### Encontrar o melhor nº de intervalos
Consideremos que usamos a Regra do Trapézio com $N_i$ intervalos e com $h_i = (b-a)/N_i$, lembrando que $N_{i-1} = \frac{1}{2}N_i$ e $h_{i-1} = 2h_i$. Então:

\
\begin{align}
I_i &= h_i \left[ \frac{f(a)}{2} + \frac{f(b)}{2} + \sum_{k=1}^{N_i-1} f(a+kh_i) \right]\\ &= h_i \left[ \frac{f(a)}{2} + \frac{f(b)}{2} + \sum_{k\,par}^{N_i-2} f(a+kh_i) + \sum_{k\,ímpar}^{N_i-1} f(a+kh_i) \right]
\end{align}

Verificamos no entanto que:

\
\begin{align}
\sum_{k\,par}^{N_i-2}f(a+2kh_i) = \sum_{k=1}^{N_i/2 - 1} = \sum_{k=1}^{N_{i-1}-1}f(a+kh_{i-1}
\end{align}

O que implica:

\begin{align}
I_i = \frac{h_{i-1}}{2} \left[ \frac{f(a)}{2} + \frac{f(b)}{2} + \sum_{k=1}^{N_{i-1}-1} f(a+kh_{i-1}) \right] + h_i \sum_{k\, ímpar}^{N_i-1}f(a+kh_i)
\end{align}

Nota: o primeiro termo é precisamente o $I_{i-1}$ dado pela regra do Trapézio, por isso...

\
\begin{align}
I_i = \frac{I_{i-1}}{2} + h_i \sum_{k\, ímpar}^{N_i-1}f(a+kh_i)
\end{align}

Ou seja, o integral $I_i$ dado por $N_i$ intervalos não precisa de ser calculado do zero, tornando o processo computacional mais rápido e preciso, pois podemos apenas necessitamos calcular uma parte, sendo que a outra é dada pelo integral anterior, $I_{i-1}$ e por $N_{i-1}$ intervalos

### Equivalente para a Regra de Simpson
\
O erro é dado por: $\epsilon_i = \frac{(I_i - I_{i-1})}{15}$; logo:
\begin{align}
&S_i = \frac{1}{3} \left[ f(a) + f(b) + 2 \sum_{k\,par}{N_i-2} f(a+kh_i) \right]\\ \\&T_i = \frac{2}{3} \sum_{k\,ímpar}^{N_i-1} f(a+kh_i)\\ \\&\Rightarrow S_i = S_{i-1} + T_{i-1} \land I_i = h_i(S_i + 2T_i) 
\end{align}

## Integração de Romberg

Voltando à regra do Trapézio, o erro é da ordem $ch_i^2 = \frac{(I_i - I_{i-1})}{3}$, sendo que por definição ,$I = I_i + ch_i^2 + O(h_i^4) = I_i + \frac{1}{3}(I_i-I_{i-1}) + O(h_i^4)$

Vamos dizer que:
\
\begin{align}
&R_{i,1} = I_i\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\,\, R_{i,2}=I_i + \frac{1}{3}(I_i - I_{i-1}) = R_{i,1} + \frac{1}{3}(R_{i,1} - R_{i-1,1})\\ \\&\Rightarrow I = R_{i,2} + c_2h_i^4 + O(h_i)^6 = R_{i-1,2} + c_2h_{i-1}^4 + O(h_{i-1}^6) = R_{i-1,2} + 16c_2h_i^4 + O(h_i^6)
\end{align}

Daqui tiramos que:
\
\begin{align}
c_2h_i^4 = \frac{1}{15}(R_{i,2} - R_{i-1,2}) + O(h_i^6)\,\,\,\,\, \Rightarrow\,\,\,\,\,I = R_{i,2} + \frac{1}{15}(R_{i,2} - R_{i-1,2}) + O(h_i^6)
\end{align}

Em suma:

\
\begin{align}
I &= R_{i,m+1} + O(h_i^{2m+2})\\ R_{i,m+1} &= R_{i,m} + \frac{1}{4^m-1}(R_{i,m} - R_{i-1,m})
\end{align}

Lembrando que $R_{i,1} = I_i$.

```python
#Algoritmo ideal para Romberg

def Romberg(f,a,b,N):
    R = np.zeros((N, N))
    R[0,0] = Trapezio(f,a,b,N)
    tol = 1e-6
    erro = 1 
    i = 1
    while erro>tol:
        N*=2
        R[i,0]=Trapezio(f,a,b,N)
        for j in range(1,i+1):
            R[i,j] = (R[i,j-1] + 1/(4**(j)-1) * (R[i,j-1] - R[i-1,j-1]))
        erro = abs(R[i,i] - R[i-1,i-1])
        i+=1
    return R[i-1,i-1] , erro , i , N

#Exemplo:
def f(x):
    return x**4 - 2*x + 1
#Resultado:
print(Romberg(f,0,2,100))
```

## Quadratura Gaussiana

Vamos supor que temos $N$ pontos $x_k$, vamos então calcular o integral de $\int_{a}^{b} f(x)\,dx \approx \sum_{k=1}^{N}w_kf(x_k)$ baseando-nos nos pontos $f(x_k)$ e usando pesos $w_k$.

Para obter estes $N$ pontos, necessitamos um polinómio de grau $N-1$, dado por um polinómio de interpolação:

\
\begin{align}
\phi_k(x) &= \prod_{m=1\, \land\, m \neq k}^{N} \frac{(x-x_m)}{(x_k-x_m)}\\ \\&=\frac{(x-x_1)}{(x_k-x_1)}\, \times\, (...)\, \times\, \frac{(x-x_{k-1})}{(x_k-x_{k-1})}\, \times\, \frac{(x-x_{k+1})}{(x_k-x_{k+1})}\, \times\, (...)\, \times\, \frac{(x-x_{N})}{(x_k-x_{N})}
\end{align}

Resumindo:

\
\begin{align}
\phi_k(x_m) = \delta _{km} =\left\{
    \begin{array}{ll}
        1 & \text{se } m=k \\
        0 & \text{se } m \neq k
    \end{array}
\right.
\end{align}

O $\delta_{km}$ é o delta de Knronecker.\\

É importante saber que este polinómio \Phi(x) de grau $N-1$ é único para $N$ pontos de $f(x)$ e é uma combinação linear de polinómios também de grau $N-1$ como veremos a seguir:

\
\begin{align}
\Phi(x_m) = \sum_{k=1}^{N} f(x_k)\phi_k(x) = \sum_{k=1}^{N} f(x_k) \delta_{km} = f(x_m)
\end{align}

Para calcular a aproximação do nosso integral:

\
\begin{align}
\int_{a}^{b} f(x) \, dx \approx \int_{a}^{b} \phi(x) \, dx = \int_{a}^{b} \sum_{k=1}^{N} f(x_k) \phi_k(x) \,dx = \sum_{k=1}^{N} f(x_k) \int_{a}^{b} \phi_k(x) \,dx
\end{align}

Note-se que $\int_{a}^{b} \phi_k(x) \,dx = w_k$ chamamos de pesos.

Além disso, poderemos ter um outro problema ao integrar numéricamente a função, através de pesos e zeros, ou seja, através da quadratura de Gauss: os limites de integração. Por norma, temos $w_k = \int_{-1}^{1} \phi_k(x) \,dx$. Mas podemos mudar os limites de integração da seguinte forma:

\
\begin{align}
x_k' &= \frac{1}{2}(b-a)x_k + \frac{1}{2}(b+a)\\
w_k' &= \frac{1}{2}(b-a)w_k
\end{align}

Daí tiramos que:

\
\begin{align}
\int_{a}^{b} f(x)\, dx \approx \sum_{k=1}^{n}w_k'f(x_k')
\end{align}

```python
#Algortimo de exemplo para a Quadratura de Gauss-Legendre:

def f(x):
    return x**2

def gauss(f,a,b,N):
    zeros, pesos = scipy.special.roots_legendre(N)
    zeros = 0.5*(b-a)*zeros + 0.5*(b+a)
    pesos = 0.5*(b-a)*pesos
    I = np.sum(f(zeros)*pesos)
    return I

print(gauss(f,-1,1,1000))

#Algortimo de exemplo para a Quadratura de Gauss-Hermite:

def hermite(f,N):
    zeros, pesos = scipy.special.roots_hermite(N)
    I = np.sum(f(zeros)*pesos)
    return I

print(hermite(f,100000))

#Algortimo de exemplo para a Quadratura de Gauss-Chebyshev:

def chebyshev(f,a,b,N):
    zeros, pesos =  scipy.special.roots_chebyt(N)
    x = 0.5*(b-a)*zeros + 0.5*(b+a)
    pesos = 0.5*(b-a)*pesos
    I = np.sum(f(x)*pesos*np.sqrt(1-zeros**2))
    return I
    
print(chebyshev(f,-1,1,1000))
```

### Eliminação de Gauss

Este é um método no qual procuramos transforma a nossa matriz $\textbf{A}$ numa matriz triangular superior:
\
\
\begin{align}
Matriz\,\,\,\,triangular\,\,\,\,superior =  \begin{bmatrix}
a_{11} & a_{12} & a_{13} & \cdots & a_{1n} \\
0 & a_{22} & a_{23} & \cdots & a_{2n} \\
0 & 0 & a_{33} & \cdots & a_{3n} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0 & 0 & 0 & \cdots & a_{nn} \\
\end{bmatrix}
\end{align}

Note-se que:
\
\begin{align}
x_n = \frac{b_n}{a_{n,n}} \implies x_{n-1} = \frac{b_{n-1}-a_{n-1,n}x_n}{a_{n-1,n-1}} \implies\,\,\, ...\,\,\, \implies x_{n-i}
= \frac{b_{n-i} - \sum_{i-1}^{j=0}\, a_{n-i,n-j}x_{n-j}}{a_{n-i,n-i}},\,\,\,\, i\leq i \leq n-1
\end{align}

```python
#Algoritmo de resolução de matrizes triangulares superiores n x n

def Resolve_tri_sup (A,v):
    n=len(v)
    x=np.zeros(n)
    x[n-1]=v[n-1]/A[n-1][n-1]
    for i in range(n-1,0,-1):
        s=0
        for j in range(i+1,n+1,1):
            s= s + A[i-1][j-1]*x[j-1]
        x[i-1]= (v[i-1]-s)/A[i-1][i-1]
    return x

A=[[3,2,4],[0,1/3,2/3],[0,0,-8]]
v=[1, 5/3,0]

print(Resolve_tri_sup(A,v))
print(np.linalg.solve(A, v))
```

```python
#1º Código da eleminação de Gauss:

def resolve_gauss(A, v):
    N = len(v)
    x = np.zeros(N)
    for p in range(N):
        for i in range(p+1, N):
            coef = A[i,p]/A[p,p]
            v[i] -= v[p] * coef
            for j in range(p, N):  # Corrigindo o índice de início do loop
                A[i,j] -= coef * A[p,j]
        print("Pivo: ", p,"\n Matriz: ", A, "\n")

    for p in range(N-1, -1, -1):
        x[p] = v[p]
        for i in range(p+1, N):
            x[p] -= A[p,i]*x[i]
        x[p] = x[p]/A[p,p]
    return x

A = np.array([[0.1,1,2,4.1],[3,4.1,5,2], [6,7,8,5], [0.5,0,3,5]])
v = np.array([0.1,0.2,0.3,0.4])
x = resolve_gauss(A.copy(), v.copy())

print("Solução x:", x)
print(np.linalg.solve(A, v))
```

### Eliminação Gauss-Jordan

```python
#1º código para eliminação Gauss-Jordan
def gauss_jordan(A,v):
    N = len(v)
    for p in range(N):
        v[p] /= A[p,p]
        A[p,p:] /= A[p,p]
        for i in range(p):
            v[i] -= v[p]*A[i,p]
            A[i,p:] -= A[p,p:] * A[i,p]
        for i in range(p+1, N):
            v[i] -=v[p] * A[i,p]
            A[i,p:] -= A[p,p:] * A[i,p]
        print("Pivot: " ,p,"\n Matirz: ", A)
    return v

A = np.array([[0.1,1,2,4.1],[3,4.1,5,2], [6,7,8,5], [0.5,0,3,5]])
v = np.array([0.1,0.2,0.3,0.4])
x = gauss_jordan(A.copy(), v.copy())

print("Solução x:", x)
print(np.linalg.solve(A, v))
```

```python
#2º código para a eliminação de Gauss-Jordan

def gauss_jordan_2(A,v):
    N = len(v)
    A_new = np.zeros([N , N+1])
    A_new[:, :N] = A
    A_new[:, N] = v
    
    for p in range(N):
        A_new[p,p:] /= A_new[p,p]
        
        for i in range(p):
            A_new[i,p:] -= A_new[p,p:] * A_new[i,p]
        for i in range(p+1, N):
            A_new[i,p:] -= A_new[p,p:] * A_new[i,p]
    return A_new[:, -1]

A = np.array([[0.1,1,2,4.1],[3,4.1,5,2], [6,7,8,5], [0.5,0,3,5]])
v = np.array([0.1,0.2,0.3,0.4])
x = gauss_jordan_2(A.copy(), v.copy())

print("Solução x:", x)
print(np.linalg.solve(A, v))
```

### Eliminação de Gauss-Jordan com recuso a pivotagem parcial

```python
#1º código gauss-jordan com pivotagem parcial e com recurso a um array para a troca das linhas
def gauss_jordan_pivparcial(A,b):
    N = len(b)
    linhas = np.arange((N))
    AA = np.zeros((N, N+1))
    AA[:N ,:N] = A
    AA[:,N]=b
    
    for p in range(N):
        indice = np.argmax(np.abs(AA[linhas[p:], p])) + p
        linhas[p], linhas[indice] = linhas[indice], linhas[p]
        AA[linhas[p], : ] /= AA[linhas[p], p]
        
        for i in range(p):
            AA[linhas[i], p:] -= AA[linhas[p], p:]*AA[linhas[i], p]
        
        for i in range(p+1,N):
            AA[linhas[i], p:] -= AA[linhas[p], p:]*AA[linhas[i], p]
        print("\nArray linhas (troca de linhas):\n", linhas,"\n Matriz A:\n", A)
        
    return AA[linhas[:], -1]

A = np.array([[0.1,1,2,4.1],[3,4.1,5,2], [6,7,8,5], [0.5,0,3,5]])
v = np.array([0.1,0.2,0.3,0.4])
x = gauss_jordan_pivparcial(A.copy(), v.copy())

print("Solução x:", x)
print(np.linalg.solve(A, v))
```

### Elimação Gauss Jordan com pivotagem total

```python
def gauss_jordan_pivtotal(A,b):
    N = len(b)
    linhas = np.arange(N)
    colunas = np.arange(N+1)
    AA = np.zeros((N,N+1))
    AA[:N, :N] = A
    AA[: , N] = b
    aa = np.zeros(N*N)
    
    for p in range(N):
        m = N-p
        for i in range(m):
            aa[i * m:(i+1) * m] = AA[linhas[p+i], colunas[p:-1]]
        indice = np.argmax(np.abs(aa[:m * m]))
        l = p + indice//m
        c = p + indice % m 
        linhas[p], linhas[l] = linhas[l], linhas[p]
        colunas[p], colunas[c] = colunas[c], colunas[p]
        AA[linhas[p], colunas[:]] /= AA[linhas[p], colunas[p]]
        for i in range(p):
            AA[linhas[i], colunas[p:]] -= AA[linhas[p], colunas[p:]]*AA[linhas[i], colunas[p]]
        for i in range(p+1, N):
            AA[linhas[i], colunas[p:]] -= AA[linhas[p], colunas[p:]]*AA[linhas[i], colunas[p]]
    return AA[colunas[linhas[ : ]] , -1]
    
A = np.array([[0.1,1,2,4.1],[3,4.1,5,2], [6,7,8,5], [0.5,0,3,5]])
v = np.array([0.1,0.2,0.3,0.4])
x = gauss_jordan_pivtotal(A.copy(), v.copy())

print("Solução x:", x)
print(np.linalg.solve(A, v))
```

### Eliminação de Gauss com pivotamento parcial

O pivotamento parcial consiste na troca entre linhas de forma a favorecer a eliminação de Gauss. Para o último método referido necessitamos que  a matriza $\textbf{A}$ seja uma triangular superior, ou seja, temos que:

\
\begin{align}
a_{i,j} \neq 0 ,\,\,\, i \leq j\,\,  \land\,\, a_{i,j} = 0 ,\,\,\, i > j
\end{align}

Mas, por vezes, o que acontece é que temos elementos $a_{i,j} = 0$ com $i \leq j$ o que nos impossibilita de implementar a eliminação de Gauss. Daí a troca de linhas, ou seja, o pivotamento parcial.

```python
#1º Código para eliminação de Gauss com pivotamento parcial

A=np.array([[1,2,1],[3,-2,-2],[1,3,-1]],float)
v=np.array([1,-1,-9],float)

def eli_gaus_compiv(A,v):
    
    N=len(v)
    x = np.empty(N, float)
    
    for m in range(N):
        
        # Encontre o índice do elemento com o maior valor absoluto na coluna atual
        k = np.argmax(np.abs(A[m:, m])) + m
        
        # Troque as linhas para colocar o elemento pivot na diagonal principal
        A[[m, k], :] = A[[k, m], :]
        v[[m, k]] = v[[k, m]]
        
        # Eliminação gaussiana
        for i in range(m+1, N):
            factor = A[i, m] / A[m, m]
            A[i, :] -= factor * A[m, :]
            v[i] -= factor * v[m]
        
        
    # Substituição retroativa
    for m in range(N-1, -1, -1):
        x[m] = v[m]
        for i in range(m+1, N):
            x[m] -= A[m, i] * x[i]
        x[m] /= A[m, m]
    return x

print(eli_gaus_compiv(A,v))
print(np.linalg.solve(A, v))
```

### Decomposição LU

Um método alternativo à eliminação de gauss é a decomposição da matriz de $\textbf{A}$ em:

\
\begin{align}
\textbf{A} =\textbf{L}\,\textbf{U}
\end{align}

- $\textbf{L}$ é uma matriz triangular inferior (zeros em cima, valores em baixo);

- $\textbf{U}$ é uma matriz triangular superior (valores em cima, zeros em baixo).

Consideremos a matriz $\textbf{A}$:

\
\begin{align}
\textbf{A} =
\begin{bmatrix}
a_{1,1} & a_{1,2} & a_{1,3} & a_{1,4} \\
a_{2,1} & a_{2,2} & a_{2,3} & a_{2,4} \\
a_{3,1} & a_{3,2} & a_{3,3} & a_{3,4} \\
a_{4,1} & a_{4,2} & a_{4,3} & a_{4,4}
\end{bmatrix}
\end{align}

Agora iremos fazer as seguintes operações:

\
\begin{align}
L_1 = \frac{L_1}{a_{1,1}} - a_{1,1} L_1
\end{align}

\
\begin{array}{cc}
\frac{1}{a_{1,1}}\,
\begin{bmatrix}
1 & 0 & 0 & a_0 \\
-a_{2,1} & a_{1,1} & 0 & 0 \\
-a_{3,1} & a_{3,2} & a_{1,1} & 0 \\
-a_{4,1} & a_{4,2} & 0 & a_{1,1}
\end{bmatrix} & 
\begin{bmatrix}
a_{1,1} & a_{1,2} & a_{1,3} & a_{1,4} \\
a_{2,1} & a_{2,2} & a_{2,3} & a_{2,4} \\
a_{3,1} & a_{3,2} & a_{3,3} & a_{3,4} \\
a_{4,1} & a_{4,2} & a_{4,3} & a_{4,4}
\end{bmatrix}=
\begin{bmatrix}
1 & b_{1,2} & b_{1,3} & b_{1,4} \\
0 & b_{2,2} & b_{2,3} & b_{2,4} \\
0 & b_{3,2} & b_{3,3} & b_{3,4} \\
0 & b_{4,2} & b_{4,3} & b_{4,4}
\end{bmatrix}
\end{array}


Desta forma denominaremos a seguinte matriz triangular inferior:

\
\begin{array}{cc}
L_1 =
\frac{1}{a_{1,1}}\,
\begin{bmatrix}
1 & 0 & 0 & a_0 \\
-a_{2,1} & a_{1,1} & 0 & 0 \\
-a_{3,1} & a_{3,2} & a_{1,1} & 0 \\
-a_{4,1} & a_{4,2} & 0 & a_{1,1}
\end{bmatrix}
\end{array}

E agora voltamos a realizar o mesmo processo, dividindo todos os elementos da matriz resultante do última operação pelo segundo elemento da diagonal:

\
\begin{array}{cc}
\frac{1}{b_{2,2}}\,
\begin{bmatrix}
b_{2,2} & 0 & 0 & 0 \\
0 & 1 & 0 & 0 \\
0 & -b_{3,2} & b_{2,2} & 0 \\
0 & -b_{4,2} & 0 & b_{2,2}
\end{bmatrix} & 
\begin{bmatrix}
1 & b_{1,2} & b_{1,3} & b_{1,4} \\
0 & b_{2,2} & b_{2,3} & b_{2,4} \\
0 & b_{3,2} & b_{3,3} & b_{3,4} \\
0 & b_{4,2} & b_{4,3} & b_{4,4}
\end{bmatrix}=
\begin{bmatrix}
1 & c_{1,2} & c_{1,3} & c_{1,4} \\
0 & 1 & c_{2,3} & c_{2,4} \\
0 & 0 & c_{3,3} & c_{3,4} \\
0 & 0 & c_{4,3} & c_{4,4}
\end{bmatrix}
\end{array}

Logo, consideramos que $L_2$ é:

\
\begin{array}{cc}
L_1=\frac{1}{b_{2,2}}\,
\begin{bmatrix}
b_{2,2} & 0 & 0 & 0 \\
0 & 1 & 0 & 0 \\
0 & -b_{3,2} & b_{2,2} & 0 \\
0 & -b_{4,2} & 0 & b_{2,2}
\end{bmatrix}
\end{array}

E com isto temos ainda:

\
\begin{array}{cc}
L_3=\frac{1}{c_{3,3}}\,
\begin{bmatrix}
c_{3,3} & 0 & 0 & 0 \\
0 & c_{3,3} & 0 & 0 \\
0 & 0 & 1 & 0 \\
0 & 0 & -c_{4,3} & c_{3,3}
\end{bmatrix}
\end{array}

\
\begin{array}{cc}
L_4=\frac{1}{d_{4,4}}\,
\begin{bmatrix}
d_{4,4} & 0 & 0 & 0 \\
0 & d_{4,4} & 0 & 0 \\
0 & 0 & d_{4,4} & 0 \\
0 & 0 & 0 & 1 
\end{bmatrix}
\end{array}

Daqui tiramos que: $L_4\,L_3\,L_2\,L_1\,A\,x=L_4\,L_3\,L_2\,L_1\,v$

\
E que $L = L_{4}^{-1}\,L_{3}^{-1}\,L_{2}^{-1}\,L_{1}^{-1}$ e $U = L_4\,L_3\,L_2\,L_1\,A$

\
Logo como $L\,U = A \implies L\,U\,x = v$ 

Mas como é que calculamos os elementos de L e U de forma efetiva? Chegamos à seguinte conclusão:

\
\begin{align}
L_{i,j} &= A_{i,j} - \sum_{k=1}^{i-1} L_{i,k}\, U_{k,j}\\ U_{i,j} &= \frac{1}{L_{i,i}}(A_{i,j} - \sum_{k=1}^{j-1} L_{i,k}\, U_{k,j})
\end{align}


```python
#Código para decomposição A em LU sem pivotamento

A=np.array([[1,2,1],[3,-2,-2],[1,3,-1]],float)
v=np.array([1,-1,-9],float)


def lu_decomposition(A):
    N = len(A)
    L = np.zeros((N, N))
    U = np.zeros((N, N))

    for j in range(N):
        L[j][j] = 1.0

        for i in range(j + 1):
            U[i][j] = A[i][j] - sum(U[k][j] * L[i][k] for k in range(i))
        
        for i in range(j, N):
            L[i][j] = (A[i][j] - sum(U[k][j] * L[i][k] for k in range(j))) / U[j][j]
    
    return L, U

L,U = lu_decomposition(A)
print(np.dot(L,U))
print(lu_decomposition(A))


A=np.array([[1,2,1],[3,-2,-2],[1,3,-1]],float)
v=np.array([1,-1,-9],float)

P,L,U = scipy.linalg.lu(A)
print(np.dot(L,U))
print(scipy.linalg.lu(A))
```

## Resolução de equações não lineares

### Método de relaxação
Peguemos na seguinte equação: $x = 2 - e^{-x} \iff x + e^{-x} = 2$
Se soubermos em que intervalo podemos encontrar a solução, então podemos determinar o resultado, por meio de iterações sucessivas. Por exemplo, peguemos no valor $x=1$:

\
$$
\begin{align}
x' = 2 - e^{-1} \approx 1,632 \implies x'' = 2 - e^{-1,632} \approx 1,804 \implies (...)
\end{align}
$$

##### Problemas deste método:
 - Nem sempre temos uma equação: $x = f(x)$; Exemplo: $\log{x} + x^2 - 1 = 0$ .
 - A equação poderá ter mais que 1 solução, não convergindo para um único valor.
 - Às vezes, a forma como etsá disposta a equação matemática poderá influenciar a convergência para a soluçãpo pretendida. Exemplo: $x = e^{1-x^2} \to$ não existe convergência para a solução que é $x=1$; já com um rearranjo da equação $ x = \sqrt{1 - \log x}$ já obtemos a solução esperada.
 
##### Teoria por trás do rearranjo de equações

Supondo que temos $f(x) = x$ e que a solução é $x = x^*$, façamos uma expansão em série de Taylor, ao valor $x'$, obtido a partir da 1ª iteração com o valor $x$:
$$
\begin{align}
x' = f(x) = f(x^*) +f'(x^*)(x-x^*) + O(x^3)
\end{align}
$$

\
Por definição $x^*$ é a solução da equação, ou seja, $f(x^*) = x^*$ :

$$
x' - x^* = f'(x^*)(x-x^*)
$$

Note-se que a distância entre o nosso $x$ "atual" e anossa solução, $(x - x^*)$ é multilpliado por um fator, $C = f'(x^*)$, logo o nosso x' só se aproximará ou convergirá para a nossa solução, se e apenas se, $C =|f'(x^*)| < 1$.

###### E se $C = |f'(x^*)|>1$ o que fazemos?
Neste caso, rearranjamos a equação, invertendo a função tal que, $x = f^{-1}(x)$. Agora calculamos a derivada:
$$
u = f^{-1}(x) \to x = f(u) \implies \frac{dx}{du} = f'(u)
$$

\
Quando $x = x^* \implies u = f^{-1}(x^*) = x^*$ , logo:
$$
\frac{du}{dx} = \frac 1 {f'(x^*)} \implies C = \frac 1{f'(x^*)} < 1
$$

#### Taxa de convergência
Vamos definir o erro em relação à iteração "atual" como $\epsilon = x^* - x$ e ao erro associado à iteração seguinte $\epsilon' = x^*- x'$, logo:
$$
\epsilon' = \epsilon f'(x^*) \implies x^* = x + \epsilon = x + \frac{\epsilon'}{f'(x^*)}
$$

\
Nisto obtemos: $\epsilon' = \frac{x-x'}{1-1/f'(x^*)} \approx \frac{x-x'}{1-1/f'(x)}$ , considerando que $x$ está muito próximo de $x^* \implies f'(x) = f'(x^*)$


### Método de relaxação para 2 ou + variáveis

#### Algoritmo de Jacobi
- Recorre ao método de relaxação.

```python
#Algoritmo de Jacobi

def jacobi(A, v, eps, x0):
    N = len(A)
    C_inv = np.linalg.inv(A*np.eye(N))
    B = np.eye(N) - C_inv@A
    d=C_inv@v
    x_old = np.copy(x0)
    count = 0
    erro = 1
    while (erro >= eps):
        x = B@x_old + d
        erro = np.max(np.abs(x - x_old))/np.max(np.abs(x))
        x_old = np.copy(x)
        count+=1
    return x

x0 = np.zeros(N)
A = [[10,2,1], [1,5,1],[2,3,10]]
v = [7,-8,6]
print(jacobi(A,v,1e-6,x0))
```

### Método da bisseção

```python
#Código para o método da bisseção

def f(x):
    return 2*np.sqrt(x)-2*np.tanh(x)-2

def bissecao(a, b, f, eps):
    if f(a) * f(b) >= 0:
        raise ValueError("A função não muda de sinal nos pontos a e b")

    x_1 = a
    x_2 = b
    x_linha = 0.5 * (x_1 + x_2)
    erro = abs(x_2 - x_1)
    while erro > eps:
        x_linha = 0.5 * (x_1 + x_2)
        if f(x_linha) * f(x_1) < 0:
            x_2 = x_linha
        else:
            x_1 = x_linha
        erro = abs(x_2 - x_1)
    return x_linha

print(bissecao(1, 10, f, 0.002197))
```

```python
def gauss_jordan_inverse(A):
    n = len(A)
    A_inv = np.zeros((n, n))
    A_ext = np.hstack((A, np.identity(n)))
    
    pivots = []

    for i in range(n):
        # Pivoteamento parcial
        pivot_row = max(range(i, n), key=lambda j: abs(A_ext[j, i]))
        pivots.append(pivot_row)
        A_ext[[i, pivot_row]] = A_ext[[pivot_row, i]]

        # Normalizando o pivô
        pivot = A_ext[i, i]
        A_ext[i] /= pivot

        # Eliminação gaussiana
        for j in range(n):
            if i != j:
                factor = A_ext[j, i]
                A_ext[j] -= factor * A_ext[i]

    # Extraindo a matriz inversa
    A_inv = A_ext[:, n:]

    return A_inv, pivots

# Testando a função
A = np.array([[2, 1, 1], [1, -1, 0], [3, -1, 2]])
A_inv, pivots = gauss_jordan_inverse(A)

print("Índices dos pivôs:")
print(pivots)
print("\nInversa de A:")
print(A_inv)

# Verificando a inversa: A * A_inv deve ser uma matriz identidade
identity = np.dot(A, A_inv)
print("\nVerificação da inversa (A * A_inv):")
print(identity)
```

```python
def f(x):
    return x**2 - 4

# Derivada da função de exemplo
def df(x):
    return 2 * x

# Método do gradiente (Gauss-Newton)
def gauss_newton(x0, learning_rate, tolerance):
    x = x0
    while True:
        grad = df(x)
        x_new = x - learning_rate * grad
        
        # Verifica a convergência
        if np.abs(f(x_new) - f(x)) < tolerance:
            break
        
        x = x_new
    
    return x

# Exemplo de uso
x0 = 2.0  # Valor inicial
learning_rate = 0.1  # Taxa de aprendizado
tolerance = 1e-6  # Tolerância para a convergência

min_x = gauss_newton(x0, learning_rate, tolerance)
print("Mínimo local encontrado:", min_x)
print("Valor da função no mínimo local:", f(min_x))
```

### Matrizes banda

```python

def matrix_tridiagonal(n):
    # Elementos da diagonal principal
    diag_principal = np.ones(n)
    
    # Elementos da subdiagonal inferior
    subdiagonal_inferior = np.full(n - 1, 1.0)
    
    # Elementos da subdiagonal superior
    subdiagonal_superior = np.full(n - 1, 1.0)
    
    # Construindo a matriz utilizando a função np.diag para a diagonal principal e np.diagflat para as subdiagonais
    A = np.diag(diag_principal) + np.diagflat(subdiagonal_inferior, -1) + np.diagflat(subdiagonal_superior, 1)
    
    return A

n = 10
A = matrix_tridiagonal(n)
print("Matriz A:")
print(A)

# Calculando o produto da matriz tridiagonal A pela matriz identidade I
I = np.eye(n)
result = np.dot(A, I)
print("\nProduto de A por I:")
print(result)
```
